```python
'''
    We want to product this dict:
        productList = [
            {
                'productId': item.get_attribute('data-productid'),
                'productName': '',
                'imageUrl': '',
                'productUrl': '',
                'description': '',
                'priceInfo': '',
                'stars': 4.5,
                'numReviews': 4500
            } 
            for item in product_grid_items
        ]
        
        
        <div class="sparky-c-card__body sparky-c-product-card__body">
          <div class="sparky-c-product-card__content">
            <div class="sparky-c-sparky-text-wrapper">
              <h2 class="sparky-c-heading sparky-c-product-card__title sparky-c-heading--body-bold">
                <a class="sparky-c-text-link sparky-c-product-card__text-link" 
                    href="/dog/treats/dental-treats-and-chews/greeniesandtrade-adult-regular-dog-dental-treats--natural-oral-health-original-4750.html" 
                    target="_self" tabindex="-1"
                >
                  Greenies™ Adult Regular Dog Dental Treats - Natural, Oral Health, Original
                </a>
              </h2>
            </div>
          </div>
        </div>
        
        <div class="sparky-c-star-rating sparky-c-product-card__star-rating sparky-c-star-rating--sm">
            <div class="sparky-c-star-rating__icons" aria-label="4.5 out of 5 stars selected">
              <svg ... /svg>
              <svg ... /svg>
              <svg ... /svg>
            </div>
            <div class="sparky-c-star-rating__rating-after" aria-label="(73258) reviews">(73258)</div>
          </div>
'''

product_dict = {} # alternatively use a constructor like: dict()
for item in product_grid_items:
    # get product id as an attribute
    productId = item.get_attribute('data-productid')
    
    # create the item details dictionary
    item_details = {}
    
    # locate values inside of the item (WebElement)    
    item_card_header = item.find_element(By.CSS_SELECTOR, 'header')
    item_card_header_img = item_card_header.find_element(By.TAG_NAME, 'img')
    
    item_card_body = item.find_element(By.CSS_SELECTOR, 'div.sparky-c-product-card__content')
    item_card_body_a_elem = item_card_body.find_element(By.TAG_NAME, 'a')
    
    item_card_footer = item.find_element(By.TAG_NAME, 'footer')
    item_price_container = item_card_footer.find_element(By.CSS_SELECTOR, 'div.sparky-c-price.sparky-c-product-card__price-group')
    
    
    # add details
    item_details['productName'] = item_card_body_a_elem.text
    item_details['imageUrl'] = {'src': item_card_header_img.get_attribute('src'), 'alt': item_card_header_img.get_attribute('alt')}
    item_details['productUrl'] = item_card_body_a_elem.get_attribute('href')
    
    try:
        item_card_header_div = item_card_header.find_element(By.TAG_NAME, 'div')
        item_details['qtyInfo'] = item_card_header_div.text
    except NoSuchElementException:
      pass
    
    item_details['priceInfo'] = item_price_container.text
    
    try:
      item_ratings_container = item_card_footer.find_element(By.CSS_SELECTOR, 'div.sparky-c-star-rating')
      item_ratings_icon_container = item_ratings_container.find_element(By.CSS_SELECTOR, 'div.sparky-c-star-rating__icons')
      item_ratings_after_div = item_ratings_container.find_element(By.CSS_SELECTOR, 'div.sparky-c-star-rating__rating-after')
      item_details['stars'] = item_ratings_icon_container.get_attribute('aria-label')
      item_details['numReviews'] = item_ratings_after_div.text
      item_details['numReviews'] = int(item_details['numReviews'][1:-1]) if item_details['numReviews'] is not None else 0
    except NoSuchElementException:
      pass
    
    # add item to product list
    product_dict[productId] = item_details
    
    
product_dict
```

```python
'''
    Loop through the product dictionary and for each product get the number of quantity types and the price for each quantitity
'''

for product_id, product_details in product_dict.items():
    driver.get(product_details['productUrl'])
    product_quantity_dict = {}

    priduct_price_container = driver.find_element(By.CSS_SELECTOR, 'div.product-price')
    product_price_div = priduct_price_container.find_element(By.TAG_NAME, 'div')
    product_size_variant_container = driver.find_element(By.CSS_SELECTOR, 'div.size-variant')
    product_size_fields = product_size_variant_container.find_element(By.CSS_SELECTOR, 'div.size-variant__fields')
    product_size_label_container = product_size_fields.find_element(By.TAG_NAME, 'dl')
    product_size_label_dd = product_size_label_container.find_element(By.TAG_NAME, 'dd')

    product_sizes_fieldset = product_size_fields.find_element(By.TAG_NAME, 'fieldset')
    product_sizes_ul = product_sizes_fieldset.find_element(By.TAG_NAME, 'ul')
    product_sizes_li_list = product_sizes_ul.find_elements(By.TAG_NAME, 'li')
    size_info = {
        'count': len(product_sizes_li_list),
        'list': []
    }
    for product_size_li in product_sizes_li_list:
        product_size_li.click()
        product_size_details = {
            'size': product_size_label_dd.text,
            'price': product_price_div.text
        }
        size_info['list'].append(product_size_details)
        
    product_details['sizeInfo'] = size_info
```

```python
with open(JSON_FILE_NAME, 'w') as file:
    json.dump(product_dict, file, indent=4)
```
